## Dataset Overview

> The dataset tied to the investigation contains Loan Data information from the year 2005 to 2014. It contains around 87 variables

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loans = pd.read_csv('prosperLoanData.csv')
```

```python
var_focus = ['ListingKey','ListingNumber','ListingCreationDate','CreditGrade','Term','LoanStatus',
             'ClosedDate','BorrowerAPR','BorrowerRate',r'ProsperRating (Alpha)',r'ListingCategory (numeric)',
             'BorrowerState','Occupation','EmploymentStatus','IsBorrowerHomeowner','CurrentlyInGroup',
             'GroupKey','IncomeRange','StatedMonthlyIncome','LoanKey','LoanNumber','LoanOriginalAmount',
             'LoanOriginationQuarter','InvestmentFromFriendsCount','InvestmentFromFriendsAmount','Investors']
loans_sub  = loans[var_focus]
loans_sub['YearBorrowed'] = loans_sub['LoanOriginationQuarter'].str[3:]
```

## Visualization 1:
### How is the distribution of Monthly income for borrowers?

The plot below shows the distribution of the variable `StatedMonthlyIncome`. The distribution of this variable is normal, evidenced by the histogram's unimodal appearance. The peak of the monthly incomes comes at around the 4000 dollar mark

```python
def histplt(df, x, bin_vals,log):
    if (log == True):
        bins = 10 ** np.arange(bin_vals[0],bin_vals[1] + bin_vals[2], bin_vals[2])
        plt.xscale('log')
    else:
        bins = np.arange(bin_vals[0],bin_vals[1] + bin_vals[2], bin_vals[2])        
    plt.hist(data = df, x = x, bins = bins)
    plt.xlabel(x)
    plt.ylabel('Count');
    
histplt(loans_sub, 'StatedMonthlyIncome', [3.08,6.1,0.05], True)
ticks = [1000,3000,10000, 30000, 100000]
labels = ['{}'.format(v) for v in ticks]
plt.xticks(ticks, labels, rotation = 50);
plt.title('Distribution of StatedMonthlyIncome');
```

## Visualization 2:
### How does Homeownership and Employment status affect The loan amounts borrowed?

>The plot below represents the Loan amounts borrowed by different people of different employment statuses, grouped by whether or not they have a home. From the plot, we can deduce the fact that employed people are seen to ask for loans the most. This applies to homeowners as well as they appear to borrow more loans compared to non homeowners

```python

ax = sb.barplot(data = loans_sub, x = 'EmploymentStatus', y = 'LoanOriginalAmount', hue = 'IsBorrowerHomeowner')
ax.legend(title = 'Homeowner status')
plt.xticks(rotation = 20);
plt.title('LoanOriginalAmounts of different EmploymentStatuse grouped by their homeowner status');
```

## Visualization 3
### What is the relationship between APR, Interest Rate against the loan amount borrowed

>The plot below shows the InterestRate and `BorrowerAPR` against the origninal amount borrowed. 
From the plot we can deduce that `BorrowerRate` and `BorrowerAPR` are strongly linked and have similar distribution against the `LoanOriginalAmount`. The heatmap shows that alot of the loans fall between amounts 0 and 15000, with the most values falling where the interest rates and APR is around 0.3 and 0.3-0.4 respectively 

```python
#Interest Rate and Borrower APR against LoanAmountBorrowed

plt.figure(figsize = [20,5])
plt.subplot(1,2,1)
bins_x = np.arange(0, 0.5 + 0.04, 0.04)
bins_y = np.arange(1000, 35000 + 4000, 4000 )
plt.hist2d(data = loans_sub, x = 'BorrowerRate', y = 'LoanOriginalAmount',
           cmin = 0.5,cmap = 'viridis_r',bins = [bins_x,bins_y] )
plt.ylabel('LoanOriginalAmount')
plt.xlabel('Borrowerrate')
plt.title('LoanOriginalAmount against Borrower interest rate')
plt.colorbar()
plt.subplot(1,2,2)
bins_x2 = np.arange(0.006,0.51 + 0.04,0.04)
plt.hist2d(data = loans_sub, x = 'BorrowerAPR', y = 'LoanOriginalAmount',
          cmin = 0.5, cmap = 'viridis_r', bins = [bins_x2,bins_y])
plt.ylabel('LoanOriginalAmount')
plt.xlabel('BorrowerAPR')
plt.title('LoanOriginalAmount against Borrower APR')

plt.colorbar();
```
